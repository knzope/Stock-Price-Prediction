```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout, LSTM
```

```python
file_names=['ADANIPORTS', 'ASIANPAINT', 'AXISBANK', 'BAJAJ-AUTO', 'BAJAJFINSV',
       'BAJFINANCE', 'BHARTIARTL', 'BPCL', 'BRITANNIA', 'CIPLA',
       'COALINDIA', 'DRREDDY', 'EICHERMOT', 'GAIL', 'GRASIM', 'HCLTECH',
       'HDFC', 'HDFCBANK', 'HEROMOTOCO', 'HINDALCO', 'HINDUNILVR',
       'ICICIBANK', 'INDUSINDBK','INFY', 'IOC', 'ITC',
       'JSWSTEEL', 'KOTAKBANK', 'LT', 'MARUTI', 'MM', 'NESTLEIND',
       'NTPC', 'ONGC', 'POWERGRID', 'RELIANCE', 'SBIN',
       'SHREECEM', 'SUNPHARMA', 'TATAMOTORS',
       'TATASTEEL', 'TCS', 'TECHM', 'TITAN', 'ULTRACEMCO', 'UPL', 'VEDL',
       'WIPRO', 'ZEEL']
```

```python
from tkinter import *
root=Tk()
from tkinter import filedialog
#root.geometry("1200x500")
company=StringVar()
pred1=StringVar()

prev_win=None
def Reset():
    global prev_win

    company.set("Select Here")
    pred1.set("")
    
    try:
        prev_win.destroy()
        prev_win=None
    except AttributeError:
        pass

from tkinter import messagebox
def Exit():
    qExit=messagebox.askyesno("System","Do you want to exit the system")
    
    if qExit:
        root.destroy()
        exit()
        
def file_opener():
   input = filedialog.askopenfile(initialdir="/")
   global file_name
   file_name=input.name

def predict_price():
    company_name=company.get()
    input_file_path=file_name
    df = pd.read_csv(input_file_path)
    # create and fit the LSTM network
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(60,1)))
    model.add(LSTM(units=50))
    model.add(Dense(1))

    model.load_weights('SavedModels/'+company_name+'.h5')
    df.drop('Unnamed: 0',axis=1,inplace=True)

    x = df['Close']

    x = np.array(x)

    x = x.reshape(60,1)

    #converting dataset into x_train and y_train
    scaler = MinMaxScaler()
    scaled_data = scaler.fit_transform(x)

    new = scaled_data.reshape(1,60,1)

    y_pred = model.predict(new)

    y_pred = scaler.inverse_transform(y_pred)
    pred1.set(y_pred[0][0]);
    
    
#Tk class is used to create a root window
root.configure(background='cyan')
root.title('STOCK PRICE PREDICTION')
root.resizable(0,0)

#Headings for the GUI written at the top of GUI
w2 = Label(root, justify=LEFT, text="STOCK PRICE PREDICTION", fg="Red", bg="cyan")
w2.config(font=("Times",30,"bold italic"))
w2.grid(row=1, column=0, columnspan=2, padx=100,pady=10)
w2 = Label(root, justify=LEFT, text="Contributors: Kunal and Pranjal", fg="Blue", bg="cyan")
w2.config(font=("Times",30,"bold italic"))
w2.grid(row=2, column=0, columnspan=2, padx=100,pady=10)

#Creating Labels for the symtoms
S1Lb = Label(root, text="Select Company", fg="Black", bg="cyan")
S1Lb.config(font=("Times",15,"bold italic"))
S1Lb.grid(row=8, column=0, padx=20,pady=20, sticky=W)

company.set("Select Here")
S1 = OptionMenu(root, company,*file_names)
S1.grid(row=8, column=0,padx=120)

x = Button(root,text="Input csv File", command=lambda:file_opener(),bg="purple",fg="white",width=15)
x.config(font=("Times",15,"bold italic"))
x.grid(row=10,column=0,pady=10)


dst = Button(root, text="Predict Price",command=predict_price, bg="darkviolet",fg="white")
dst.config(font=("Times",15,"bold italic"))
dst.grid(row=8, column=1,padx=10)

#Showing the output of different aldorithms
t1=Label(root,font=("Times",15,"bold italic"),text="Predicted Price",height=1,bg="white"
         ,width=20,fg="darkviolet",textvariable=pred1,relief="sunken").grid(row=9, column=1, padx=10)

rs = Button(root,text="Reset Inputs", command=Reset,bg="red",fg="white",width=15)
rs.config(font=("Times",15,"bold italic"))
rs.grid(row=11,column=1,padx=10)

ex = Button(root,text="Exit", command=Exit,bg="red",fg="white",width=15)
ex.config(font=("Times",15,"bold italic"))
ex.grid(row=12,column=1,padx=10,pady=10)

# #calling this function because the application is ready to run
root.mainloop()
```
